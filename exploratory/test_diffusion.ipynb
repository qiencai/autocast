```python
import torch
import torch.nn as nn
from auto_cast.processors.diffusion import DiffusionProcessor, LogLinearSchedule

class SimpleUNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(3,64,3,padding=1),
            nn.ReLU(),
            nn.Conv2d(64,3,3,padding=1)
        )
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.net(x)
# Create components
denoiser_nn = SimpleUNet()
loss = nn.MSELoss()
schedule = LogLinearSchedule(sigma_min=1e-3, sigma_max=1e3)
```

```python
from auto_cast.processors.diffusion import DiffusionProcessor, LogLinearSchedule

# Create noise schedule
schedule = LogLinearSchedule(sigma_min=1e-3, sigma_max=1e3)

# Create processor
processor = DiffusionProcessor(
    denoiser_nn=denoiser_nn,
    loss=loss,
    schedule=schedule
)
```

```python
x = torch.randn(4, 3, 32, 32)  # batch=4, channels=3, height=32, width=32
output = processor(x)
print(output.shape)
```
